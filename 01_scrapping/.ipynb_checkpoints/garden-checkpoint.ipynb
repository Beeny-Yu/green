```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
from selenium.webdriver.common.keys import Keys

# options = webdriver.ChromeOptions()
# options.add_argument('--headless')
# options.add_argument('--no-sandbox')
# options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(r"C:\Users\Pokka\chromedriver.exe")
url = "https://www.modunong.or.kr:449/contents/garden/gardenList.do"
driver.get(url)

green_names = []
green_addrs = []
green_areas = []
green_ids = []


    
for i in range(3,13):
    driver.find_element_by_xpath("//*[@id='listForm']/div[2]/div[2]/ul/li["+str(i)+"]/a").send_keys(Keys.ENTER) # 페이지 넘기기    
    
    for j in range(1,13):
        green_name = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > div.tp > div.subj').text
        green_names.append(green_name)
        green_addr = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > ul > li.addr').text
        green_addr = green_addr.replace('주소','')
        green_addrs.append(green_addr)
        green_area = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > ul > li.area > p:nth-child(1)').text
        green_area = green_area.replace('전체면적','')
        green_area = green_area.replace('m²','')
        green_areas.append(green_area)
        green_id = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > div.tp > div.v_num').text
        green_id = green_id.replace('\n조회수','')
        green_ids.append(green_id)
driver.close()
        
print(green_areas)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import numpy as np
from selenium.webdriver.common.keys import Keys

# options = webdriver.ChromeOptions()
# options.add_argument('--headless')
# options.add_argument('--no-sandbox')
# options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(r"C:\Users\Pokka\chromedriver.exe")
url = "https://www.modunong.or.kr:449/contents/garden/gardenList.do"
driver.get(url)

green_lists = []

for x in range(1, 11) :
    for i in range(3, 13): # 한 번 당 10페이지씩
        driver.find_element_by_xpath("//*[@id='listForm']/div[2]/div[2]/ul/li["+str(i)+"]/a").send_keys(Keys.ENTER) # 페이지 넘기기    
        print(driver.find_element_by_css_selector("#listForm > div.cont_area > div.tbl_paging > ul > li.paging_num.on > a").text) # 현재 페이지 수 확인

        for j in range(1, 13): # 데이터 13개 for문
            try :
                green_name = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > div.tp > div.subj').text
                
                green_addr = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > ul > li.addr').text
                green_addr = green_addr.replace('주소','')
                
                green_area = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > ul > li.area > p:nth-child(1)').text
                green_area = green_area.replace('전체면적','')
                green_area = green_area.replace('m²','')
                
                green_id = driver.find_element_by_css_selector('#listForm > div.cont_area > div.plan > ul > li:nth-child('+str(j)+') > a > div.tp > div.v_num').text
                green_id = green_id.replace('\n조회수','')
                
                # 리스트에 넣기
                green_lists.append([green_id, green_name, green_addr, green_area])
            except :
                green_name = ''
                green_addr = ''
                green_area = ''
                green_id = ''
                
    # 10페이지 마다 다음 버튼 클릭
    driver.find_element_by_xpath("//*[@id='listForm']/div[2]/div[2]/ul/li[13]/a").send_keys(Keys.ENTER) # 페이지 넘기기    
driver.close()
df = pd.DataFrame(green_lists, columns=["id","텃밭명","텃밭주소","전체면적"])
df
```
