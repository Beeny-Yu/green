# 카카오 API 사용한 스크래핑 코드

### 1. 친환경 매장 : 아름다운 가게

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
import requests
import numpy as np

def location(add_searching, page_num) :
  url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
  params = {'query':add_searching, 'page' : page_num} # page가 4페이지 이상이면 15개의 결과값을 가져온다 
  headers = {"Authorization": "KakaoAK 75ac769cc8a65ecd54420549e41e515d"}

  places = requests.get(url, headers = headers, params = params).json()['documents'] # 결과 15개씩 가져옴
  cnt = requests.get(url, headers = headers, params = params).json()['meta']['total_count'] # 15개 이상이면 for문
  return places, cnt

# 결과를 저장할 리스트
store_list = []
area_name = ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주' ]
searching = '아름다운가게 사회복지단체'

for i in range(len(area_name)) :
  page_cnt = 1
  add_searching = area_name[i] + " " + searching

  # 지역별 총 total cnt 구하기
  places, cnt = location(add_searching, 1)

  if cnt <= 15 :
    page_cnt = 1
  elif cnt <= 30:
    page_cnt = 2
  else :
    page_cnt = 3
  
  for y in range(1, page_cnt+1) :
    places, cnt = location(add_searching, y)

    for place in places :
      store_list.append([place['id'], place['place_name'], place['road_address_name'], float(place['x']), float(place['y'])])

df = pd.DataFrame(store_list, columns=["id","가게명","가게주소","x좌표", "y좌표"])
# df.to_excel('beautiful_store_list.xlsx')
df
```

### 제로 웨이스트 샵 - 카카오 지도 북마크 모음
<br>크롤링해온 링크 https://map.kakao.com/?map_type=TYPE_MAP&folderid=4049789&target=other&page=bookmark

```python
import time
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import requests
import csv
import urllib.request
import urllib.parse
import re
from bs4 import BeautifulSoup
```

```python
url = "https://map.kakao.com/?map_type=TYPE_MAP&folderid=4049789&target=other&page=bookmark"

# step2
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# step3
driver = webdriver.Chrome('chromedriver',options=options)

driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
store_lists = []
# 전체 박스
box = driver.find_element_by_css_selector('#other\.favorite > ul')
tot_cnt = driver.find_element_by_css_selector('#info\.other > div.header > div > div.FavoriteDetailSummary > div.info_summary > div:nth-child(1) > a > em').text

page_cnt = int(tot_cnt) // 100 + 1 # 페이지 수
cnt = int(tot_cnt) # 결과값


for i in range(1, page_cnt + 1) :
  driver.find_element_by_css_selector("#info\.other\.favorite\.page\.no"+str(i)).send_keys(Keys.ENTER) # 페이지 넘기기
  print(i,'페이지')
  
  if cnt < 100 :
    tmp_cnt = cnt
  elif cnt > 100 :
    tmp_cnt = 100

  for x in range(1, tmp_cnt + 1) :
    try :
      # 가게명
      store_name = box.find_element_by_css_selector('li:nth-child('+str(x)+') > div.FavoriteInformationBundle > div > div > div.tit_directory > strong > a').text
      # 가게주소
      store_address = box.find_element_by_css_selector('li:nth-child('+str(x)+') > div.FavoriteInformationBundle > div > div > div.desc_directory > span').text
      # 상세정보
      store_detail = box.find_element_by_css_selector('li:nth-child('+str(x)+') > div.FavoriteInformationBundle > div > div > div.desc_directory > p.desc_detail').text
    except :
      store_detail = ''

    # 리스트에 넣기
    store_lists.append([store_name, store_address, store_detail])

  cnt -= 100 # 한 페이지 끝나면 100 빼기
```

```python
df = pd.DataFrame(store_lists, columns=["가게명","가게주소","상세정보"])
# df.to_excel('가게_제로웨이스트_3.제로웨이스트샵 카카오지도.xlsx')
df
```
