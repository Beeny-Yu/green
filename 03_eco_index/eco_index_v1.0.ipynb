# 기본 설정

```python
import pandas as pd
```

```python
# variable.csv : shop ~ car  값을 면적으로 나눈 값
df = pd.read_csv('/content/drive/MyDrive/비없비/지수/variable.csv', index_col = 0)
df
```

```python
df.shape
```

# min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler
# 객체 생성
minmax_scaler = MinMaxScaler()
# 분포 저장
minmax_scaler.fit(df)
# 스케일링
df_minmax_scaled = minmax_scaler.transform(df)
```

```python
df_minmax_scaled = pd.DataFrame(df_minmax_scaled)
df_minmax_scaled
```

```python
df_minmax_scaled.min()
```

```python
df_minmax_scaled.max()
```

# 카테고리 별 통합
- 한 카테고리 내 두개 이상의 값이 있을 때 평균 값을 지수로 계산
- 단일 값인 경우 그대로 사용

```python
# 컬럼 명 표시
df_minmax_scaled.columns = ['shop', 'park', 'trail', 'library', 'book', 'bike', 'car', 'recycle_rate', 're_center', 'vinyl', 'budget']
df_minmax_scaled.head()
```

```python
# df_new 테이블 생성
df_new = df_minmax_scaled
df_new.head(2)
```

```python
# green 행 추가
df_new['green'] = (df_new['park'] + df_new['trail']) / 2
df_new.head()
```

```python
# share 행 추가
df_new['share'] = (df_new['library'] + df_new['book']) / 2
df_new.head()
```

```python
# trans 행 추가
df_new['trans'] = (df_new['bike'] + df_new['car']) / 2
df_new.head()
```

```python
# recycle 행 추가
df_new['recycle'] = (df_new['recycle_rate'] + df_new['re_center'] + df_new['vinyl']) / 3
df_new.head()
```

```python
# 필요 없는 행 drop
df_new = df_new.drop(columns = ['park', 'trail', 'library', 'book', 'bike', 'car', 'recycle_rate', 're_center', 'vinyl'])
df_new
```

```python
df_new.min()
```

```python
df_new.max()
```

# 가중치 부여
- shop : 1.467
- green : 1.573
- share : 1.328
- trans : 1.481
- recycle : 1.732
- budget : 1.420

```python
# 기존 데이터 유지, 'df_result' 테이블 생성
df_result = df_new
```

```python
df_result.head()
```

```python
df_result['shop_rate'] = df_result['shop'] * 1.467
df_result['green_rate'] = df_result['green'] * 1.573
df_result['share_rate'] = df_result['share'] * 1.328
df_result['trans_rate'] = df_result['trans'] * 1.481
df_result['recycle_rate'] = df_result['recycle'] * 1.732
df_result['budget_rate'] = df_result['budget'] * 1.420
df_result.head()
```

```python
# 필요 없는 행 삭제
df_result = df_result.drop(columns=['shop', 'budget', 'green', 'share', 'trans', 'recycle'])
df_result
```

# 최종 점수 계산

```python
df_result['total'] = (df_result['shop_rate'] + df_result['green_rate'] + df_result['share_rate'] + df_result['trans_rate'] + df_result['recycle_rate'] + df_result['budget_rate'] ) / 6
df_result
```
